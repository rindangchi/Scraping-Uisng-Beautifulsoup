```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import pandas as pd
```

```python
#request connection ke source

#set url
URL = "https://quotes.toscrape.com/"

#get connection
resp = requests.get(URL)

#check status code
resp.status_code
```

```python
#mengubah response code ke dalam bentuk text
resp.text
```

```python
#inisiasi beautifulsoup

#establish connection
#resp = requests.get(url)

#create object for scraping
soup = BeautifulSoup(resp.text, "html.parser")

soup
```

```python
#get all data based on html class

soup.find_all("div", class_ = "quote")
```

```python
soup.find("div", class_ = "quote")
```

```python
#extract semua value dari tag quote pada website tersebut

raw_data = soup.find_all("div", class_ = "quote")

#iterate to each tag

for data in raw_data:

  #extract quote data
  quotes = data.find("span").text

  #extract author data
  author = data.find("small").text

  #get tags data
  tags = data.find_all("a" , class_ = "tag")

  #create empty list to store tags value

  get_tag = []

  for tag in tags:
    #store the tag value

    get_tag.append(tag.text)


  print(get_tag)
```

```python
# list to srore all extracted data

full_data = []

for page in tqdm(range(1,11)):
  resp = requests.get(f"https://quotes.toscrape.com/page/{page}")


  soup = BeautifulSoup(resp.text, "html.parser")


  raw_data = soup.find_all("div", class_ = "quote")

  #iterate to each tag

  for data in raw_data:

    #extract quote data
    quotes = data.find("span").text

    #extract author data
    author = data.find("small").text

    #get tags data
    tags = data.find_all("a" , class_ = "tag")

    #create empty list to store tags value
    get_tag = []

    for tag in tags:
      #store the tag value

      get_tag.append(tag.text)


    data = {
        "quote" : quotes,
        "author": author,
        "tags" : get_tag
    }

    full_data.append(data)

full_data
```

```python
len(full_data)
```

```python
quote_data = pd.DataFrame(full_data)

quote_data
```

```python
#save outout to csv

quote_data.to_csv("quote_scrape_data.csv", index= False)
```
